```python
import pandas as pd
import numpy as np

df = pd.read_csv("14.csv")
```

```python
df
```

```python
df["age"].isna().mean()*100
```

```python

df["age"] = df["age"].fillna(value=df.groupby("native.country")["age"].transform("mean"))
df["age"] = df["age"].fillna(value=df["age"].mean())
df["workclass"] = df["workclass"].fillna(value=df["workclass"].mode()[0])
df["education"] = df["education"].fillna(value=df["education"].mode()[0])
df["income"] = df["income"].fillna(value=df["income"].mode()[0])
df["occupation"] = df["occupation"].fillna(value=df.groupby("income")["occupation"].transform((lambda x: x.mode()[0] if not x.mode().empty else None)))
df["hours.per.week"] = df["hours.per.week"].fillna(value=df.groupby("occupation")["hours.per.week"].transform("median"))
df["race"] = df["race"].fillna(value=df.groupby("native.country")["race"].transform(lambda x: x.mode()[0] if not x.mode().empty else None))
df["education.num"] = df["education.num"].fillna(value=df.groupby("education")["education.num"].transform("mean"))
df["native.country"] = df["native.country"].fillna(value=df["native.country"].mode()[0])
df["capital.gain"] = df["capital.gain"].fillna(value=df["capital.gain"].mode()[0])
df["capital.loss"] = df["capital.loss"].fillna(value=df["capital.loss"].mode()[0])
df["relationship"] = df["relationship"].fillna(value=df["relationship"].mode()[0])
df["fnlwgt"] = df["fnlwgt"].fillna(value=df["fnlwgt"].mode()[0])
df['native.country'] = df['native.country'].replace("United-States",'United-Kingdom')
df["marital.status"] = df["marital.status"].fillna(value=df.groupby("relationship")["marital.status"].transform(lambda x: x.mode()[0] if not x.mode().empty else None))
def get_mode(series):
    return series.mode().iloc[0] if not series.mode().empty else None
df['sex'] = df.groupby('relationship')['sex'].transform(lambda x: x.fillna(get_mode(x)))
df.isna().sum()
```

```python
df.to_csv("clean_dataset.csv")
```
